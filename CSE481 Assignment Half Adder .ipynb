```python
"""
The half adder is able to add two single binary digits 
It has two inputs q0 and q1 and two outputs c0, c1
"""
```

```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, execute, transpile, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
qc = QuantumCircuit(4,2) 
qc.draw()
```

```python
qc.x([0,1])
qc.draw()
```

```python
qc.cx(0,2)
qc.draw()
```

```python
qc.cx(1,2)
qc.draw()
```

```python
# use ccx to write the AND of the inputs on qubit 3
qc.ccx(0,1,3)
qc.draw()
```

```python
# extract outputs
qc.measure(2,0) # extract XOR value
qc.measure(3,1) # extract AND value
```

```python
qc.draw()
```

```python
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
backend = provider.get_backend('ibm_oslo')
```

```python
qc2 = transpile(qc,backend)
qc2.draw()
```

```python
ans = backend.run(qc2,shots=800) # run in real quantum computer 800 times
counts = ans.result().get_counts()
```

```python
plot_histogram(counts)
```
